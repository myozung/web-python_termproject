## 1. 주제 및 주제 선정 이유

주제 : 교통사고 발생률 분석 

주제 선정 이유 : 자동차는 전 세계적으로 산업발전에 중추적인 역할을 하고 있지만, 동시에 교통혼잡, 환경오염, 교통사고 등의 사회문제를 발생시키고 있다. 그 중에서 사건사고를 수치적으로 가장 명확하게 확인할 수 있는 교통사고에 대해 다루어 보려고 한다. 특히 OECD의 통계자료를 보면, 우리나라는 교통사고 사망률이 낮은 선진국부터 순위를 매겼을 때 36개국 중 27위로, 사망률이 높은 하위국에 속하는 것을 확인할 수 있었다. 따라서 이러한 분석이 유의미 할 것이라고 생각한다. 

## 2. 가설 정의
1. 가해당사종이 화물차일 때 사망 사고 비율이 가장 높을 것이다.
- 화물차는 승용차/승합차/자전거/건설기계 등과 같은 다른 가해당사종에 비해 무겁고 크기 때문에 교통사고가 났을 때 사망사고로 이어질 확률이 더 높을 것이라고 생각한다.
2. 피해당사자종이 보행자일 때 사망 사고 비율이 가장 높을 것이다. 
- 피해당사종이 승용차/화물차/특수차/오토바이 등 일 때보다 보행자일 때 피해자가 입는 충격이 더 크기 때문에 (더 직접적으로 충격을 받기 때문에) 사망사고로 이어질 확률이 더 높을 것이라고 생각한다. 
3. 퇴근시간인 18-20시에 교통사고 사고건수가 가장 많을 것이다. 
- 도로에 차가 가장 많을 시간인 퇴근시간에 교통사고가 가장 많이 일어날 것이라고 생각한다.

## 4. 가설별 분석 결과

### 4.1 첫번째 가설 분석과 결과 

```python
import matplotlib.pylab as plt
from matplotlib import rc  

plt.figure(figsize = (5, 5))
plt.bar(df1["가해당사자종"], df1["사고건수"])
plt.title("가해당사자종에 따른 사고건수")
plt.xlabel("가해당사자종", fontsize = 12)
plt.ylabel("사고건수", fontsize = 12)
plt.xticks(rotation = 45, ha = "right", fontsize = 10)
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.tight_layout()

df1 = df1.sort_values(by=['사고건수'], ascending=False)

plt.show()
```

가해당사자종 별로 사고건수를 확인하기 위해서 가로축이 가해당사자종이고, 세로축이 사고건수인 bar 그래프를 그린다.

사고건수가 많은 순서대로 가해당사자종을 나열해보면(.sort_values 이용) 승용차, 화물차, 이륜차 ... 순서임을 알 수 있다.

그런데 사고건수가 많은 것과 사망 사고 비율이 높은 것은 다른 의미이기 때문에 이 자료를 통해서는 내가 세운 가설을 정확하게 확인할 수 없다. 

따라서 아래와 같은 방식으로 사망 사고 비율을 계산해서 가설을 확인한다.

```python
seung_acc = 0
seung_die = 0
seung_rate = 0

seunghab_acc = 0
seunghab_die = 0
seunghab_rate = 0

hwa_acc = 0
hwa_die = 0
hwa_rate = 0

teuk_acc = 0
teuk_die = 0
teuk_rate = 0

lee_acc = 0
lee_die = 0
lee_rate = 0 

ATV_acc = 0
ATV_die = 0
ATV_rate = 0

won_acc = 0
won_die = 0
won_rate = 0

bike_acc = 0
bike_die = 0
bike_rate = 0

PM_acc = 0
PM_die = 0
PM_rate = 0

cons_acc = 0
cons_die = 0
cons_rate = 0

agri_acc = 0
agri_die = 0
agri_rate = 0

etc_acc = 0
etc_die = 0
etc_rate = 0

for index, row in df1.iterrows():
    if row['가해당사자종'] == "승용차":
        seung_acc += int(row[2])
        seung_die += int(row[3])
    elif row['가해당사자종'] == "승합차":
        seunghab_acc += int(row[2])
        seunghab_die += int(row[3])
    elif row['가해당사자종'] == "화물차":
        hwa_acc += int(row[2])
        hwa_die += int(row[3])
    elif row['가해당사자종'] == "특수차":
        teuk_acc += int(row[2])
        teuk_die += int(row[3])
    elif row['가해당사자종'] == "이륜차":
        lee_acc += int(row[2])
        lee_die += int(row[3])
    elif row['가해당사자종'] == "사륜오토바이(ATV)":
        ATV_acc += int(row[2])
        ATV_die += int(row[3])
    elif row['가해당사자종'] == "원동기장치자전거":
        won_acc += int(row[2])
        won_die += int(row[3])
    elif row['가해당사자종'] == "자전거":
        bike_acc += int(row[2])
        bike_die += int(row[3])
    elif row['가해당사자종'] == "개인형이동장치(PM)":
        PM_acc += int(row[2])
        PM_die += int(row[3])
    elif row['가해당사자종'] == "건설기계":
        cons_acc += int(row[2])
        cons_die += int(row[3])
    elif row['가해당사자종'] == "농기계":
        agri_acc += int(row[2])
        agri_die += int(row[3])
    else :
        etc_acc += int(row[2])
        etc_die += int(row[3])
        
seung_rate = seung_die / seung_acc
seunghab_rate = seunghab_die / seunghab_acc
hwa_rate = hwa_die / hwa_acc
teuk_rate = teuk_die / teuk_acc
lee_rate = lee_die / lee_acc
ATV_rate = ATV_die / ATV_acc
won_rate = won_die / won_acc
bike_rate = bike_die / bike_acc
PM_rate = PM_die / PM_acc
cons_rate = cons_die / cons_acc
agri_rate = agri_die / agri_acc
etc_rate = etc_die / etc_acc
```

for문과 if문을 이용해서 가해당사자종마다 사고건수와 사망 사고 건수를 구하고 사망사고 비율을 계산한다.

```python
import matplotlib.pylab as plt
from matplotlib import rc 

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

car =['승용차','승합차','화물차','특수차','이륜차','사륜오토바이','원동기','자전거','개인형이동장치','건설기계','농기계','기타']
rate = [seung_rate,seunghab_rate,hwa_rate,teuk_rate,lee_rate,ATV_rate,won_rate,bike_rate,PM_rate,cons_rate,agri_rate,etc_rate]

plt.figure(figsize = (10, 5))
plt.barh(car, rate)

plt.title("가해당사자종에 따른 사망사고 비율")
plt.xlabel("사망사고비율", fontsize = 12)
plt.ylabel("가해당사자종", fontsize = 12)


plt.show()
```

가설1을 통해서 가해당사자종이 농기계일 때 사망사고 비율이 가장 높고, 가설2를 통해서 피해당사자종이 농기계일 때 사망사고 비율이 가장 높다는 것을 알 수 있었다. 가해당사자종 별, 피해당사자종 별로 분석을 했을 때 두 경우 모두 농기계일 때 사망사고 비율이 높은 이유는 일단 가설을 세울 때 생각했던 대로 농기계는 대부분 크고 높은 속도로 움직인 것 같다. 그리고 또 다른 이유로는 농업 환경에서 일하는 사람들 중에 농기계를 안전하게 사용하는 방법을 교육받은 사람이 많지 않다는 것, 또 농기계를 사용하는 곳은 대부분 땅이 포장되어 있지 않은 지역이라는 것이다. 이러한 지형 특성 때문에 사망사고가 많이 일어난다고 예측할 수 있었다. 마지막으로 농촌에서는 도시에서보다는 인프라가 덜 구축되어 있기 때문에 농기계의 유지보수가 적절하게 이루어지지 않을 것이고 따라서 사건사고가 사망사고로 이어지는 경우가 많을 것이라고 생각해볼 수 있었다. 이번 텀프로젝트를 통해서 농업 환경에서 적절한 교육, 농기계 유지보수를 위한 시설 등이 필요하다는 것을 알게 되었다.

가설3은 차가 많을 수록 사고가 많이 일어날 것이고, 퇴근시간에 도로에 차가 가장 많을 것이므로 퇴근시간에 사고가 많이 일어날 것이라는 생각으로 세웠었다. 분석 결과 가설3은 옳다는 것을 알 수 있었고, 유동적인 출퇴근 시간을 적용해서 퇴근시간에 차가 조금이라도 덜 몰리게 한다면 사고 발생을 줄일 수 있을 것이라고 생각했다.
